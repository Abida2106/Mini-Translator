```python
#MINI PROJECT 2
#MINI TRANSLATOR

import tkinter as tk
from tkinter import ttk, messagebox
from deep_translator import GoogleTranslator

# Create translator instance
translator = GoogleTranslator(source='auto', target='english')

# Get supported languages
SUPPORTED_LANGUAGES = translator.get_supported_languages(as_dict=True)

# Create dictionary with capitalized names for dropdown
LANGUAGES = {lang.title(): code for lang, code in SUPPORTED_LANGUAGES.items()}

# Function to translate text
def translate_text():
    try:
        src_text = input_text.get("1.0", tk.END).strip()
        dest_lang_name = lang_choice.get()
        dest_lang = LANGUAGES.get(dest_lang_name)

        if not src_text:
            messagebox.showwarning("Warning", "Please enter some text to translate!")
            return

        translation = GoogleTranslator(source='auto', target=dest_lang).translate(src_text)
        output_text.delete("1.0", tk.END)
        output_text.insert(tk.END, translation)

    except Exception as e:
        messagebox.showerror("Error", f"Translation failed!\n{e}")

# Tkinter setup
root = tk.Tk()
root.title("üåç Mini Translator App")
root.geometry("600x450")
root.config(bg="#f4f4f9")

# Title
title = tk.Label(root, text="üåç Mini Translator", font=("Arial", 18, "bold"), bg="#f4f4f9", fg="#333")
title.pack(pady=10)

# Input Text
tk.Label(root, text="Enter text:", font=("Arial", 12), bg="#f4f4f9").pack(anchor="w", padx=20)
input_text = tk.Text(root, height=5, width=70, font=("Arial", 12))
input_text.pack(padx=20, pady=5)

# Language Dropdown
tk.Label(root, text="Choose language:", font=("Arial", 12), bg="#f4f4f9").pack(anchor="w", padx=20)
lang_choice = ttk.Combobox(
    root,
    values=list(LANGUAGES.keys()),  # full names
    font=("Arial", 12),
    width=30
)
lang_choice.set("Hindi")  # default
lang_choice.pack(padx=20, pady=5)

# Translate Button
translate_btn = tk.Button(root, text="Translate", command=translate_text,
                          font=("Arial", 12, "bold"),
                          bg="#4CAF50", fg="white", padx=10, pady=5)
translate_btn.pack(pady=10)

# Output Text
tk.Label(root, text="Translated text:", font=("Arial", 12), bg="#f4f4f9").pack(anchor="w", padx=20)
output_text = tk.Text(root, height=5, width=70, font=("Arial", 12), fg="blue")
output_text.pack(padx=20, pady=5)

root.mainloop()
```
